#  """List of sample names.

    Returns a list of sample names in the format
    [
        "BB_940",
        "BB_941",
        "BB_943",
        "BB_944",
        "BB_945",
        "BB_946",
        "BB_947",
        ...
    ]
    """

````python
def names():
    inspector = inspect(engine)
    columns = inspector.get_columns('samples')
    sample_columns = []
    for c in columns:
        sample_columns.append(c)
    
    sample_names= []
    for names in sample_columns:
        sample_names.append(names["name"])

    samplenames = sample_names[1:]
    return jsonify(samplenames)

print (samplenames)
        
   
````

#    """MetaData for a given sample.

    Args: Sample in the format: `BB_940`

        ETHNICITY: "Caucasian",

        SAMPLEID: 940
    }
    """

````python
def metadata(metainput):
    for AGE, BBTYPE, ETHNICITY, GENDER, LOCATION, SAMPLEID in session.query(meta.AGE, meta.BBTYPE, meta.ETHNICITY, meta.GENDER, meta.LOCATION, meta.SAMPLEID).\
        filter_by(SAMPLEID = metainput):
            SAMPLEID = ["BB_" + sample for sample in SAMPLEID]
            metadict = {'AGE': AGE, 
                          "BBTY": BBTYPE, 
                          'ETHINICITY': ETHNICITY,
                          'GENDER': GENDER,
                          'LOCATION': LOCATION,
                          'SAMPLEID': SAMPLEID 
                        }
            print(SAMPLEID)
    return metadict

print(metadata("BB_940"))
    

#inspector = inspect(engine)
#columns = inspector.get_columns('samples_metadata')
#stuff = []
#for c in columns:
    #print(c['name'], c["type"])
    #stuff.append(c)
````

#     """Weekly Washing Frequency as a number.

    Args: Sample in the format: `BB_940`

    Returns an integer value for the weekly washing frequency `WFREQ`
    """
```

````python
def wash_frequency(sample_input):

    for SAMPLEID, WFREQ in session.query(meta.SAMPLEID, meta.WFREQ).\
        filter_by(SAMPLEID = sample_input):
        return WFREQ

wash_frequency("940")
````

````python
def samples_sample():
    results = session.query(samples.otu_id, samples.BB_940).\
        filter(samples.BB_940 >= 1).order_by(samples.BB_940.desc()).all()
    samplessample = [{
        "otu_ids": [result[0] for result in results],
        "sample_values": [result[1] for result in results]
        }]
    return samplessample
print(samples_sample())
````
